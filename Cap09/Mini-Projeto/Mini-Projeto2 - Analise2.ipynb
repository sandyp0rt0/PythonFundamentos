```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

## Número de veículos pertencentes a cada marca

```python
#list(df)

fig = plt.figure(figsize=[13,10])

ax = sns.countplot(y=df.brand,palette="Reds_r")
plt.xlabel("Número de Veículos")
plt.ylabel("Marca")
plt.title("Veículos por Marca")
plt.show()
```

## Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

```python
list(df)
df1 = df.groupby(["vehicleType","gearbox"])["price"].mean()
df1
```

```python
pal = dict(manuell="lime",automatik="darkorange",Unspecified="mediumpurple")
fig = plt.figure(figsize=[10,8])

sns.barplot(x="vehicleType",y="price",hue="gearbox",data=df,
            palette=pal)
            #color=["green","orange","purple"])
plt.xlabel("Tipo de Veículo")
plt.ylabel("Preço Médio")
plt.title("Preço médio dos veículos por tipo de veículo e tipo de caixa de câmbio")
plt.show()
```
