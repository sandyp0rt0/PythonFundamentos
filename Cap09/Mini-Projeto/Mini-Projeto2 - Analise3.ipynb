```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
list(df)
```

```python
pal = dict(manuell="palevioletred",automatik="olivedrab",Unspecified="steelblue")

fig = plt.figure(figsize=[8,6])

sns.barplot(x="fuelType",y="price",hue="gearbox",data=df,palette=pal)
plt.xlabel("Tipo de Combustível")
plt.ylabel("Preço Médio")
plt.title("Preço médio do veículo por tipo de combustível e tipo de caixa de cambio")
plt.show()
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
pal = dict(manuell="steelblue",automatik="goldenrod",Unspecified="silver")

fig = plt.figure(figsize=[8,6])

sns.barplot(x="vehicleType",y="powerPS",hue="gearbox",data=df,palette=pal)
plt.xlabel("Tipo de Veículo")
plt.ylabel("Potência Média")
plt.title("Potência média de um veículo por tipo de veículo e tipo de caixa de cambio")
plt.show()
```
