```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

```python
df1 = df.groupby(["brand","vehicleType"])["price"].mean()
#df1 = pd.DataFrame(df1,columns=["brand","vehicleType","avgPrice"])
df2 = df1.reset_index()
df2 = df2.pivot("brand","vehicleType","price")
df2 = df2.fillna(0)
df2 = df2.astype(int)
df2
```

```python
# Calcule a média de preço por marca e por veículo
```

## Preço médio de um veículo por marca, bem como tipo de veículo

```python
fig = plt.figure(figsize=[15,20])

sns.heatmap(df2,annot=True,fmt="d",cmap="YlGnBu")
plt.xlabel("Tipo de Veículo")
plt.ylabel("Marca")
plt.title("Heatmap - Preço médio de um veículo por marca e tipo de veículo")
plt.show()
```
