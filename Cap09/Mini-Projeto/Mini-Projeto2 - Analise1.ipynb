```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

## Distribuição de Veículos com base no Ano de Registro

```python
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro
#df.yearOfRegistration.hist(bins=len(df.yearOfRegistration.value_counts()))
#plt.title("Distribuição de Veículos com base no Ano de Registro")
#plt.xlabel("Ano")
#plt.ylabel("Quantidade")
#plt.show()

fig = plt.figure()
ax = sns.distplot(df.yearOfRegistration)
ax.set(xlabel="Ano de Registro",ylabel="Distribuição (KDE)",title=
       "Distribuição de Veículos com base no Ano de Registro")
```

## Variação da faixa de preço pelo tipo de veículo

```python
#list(df)
#df.vehicleType.value_counts()
#df.price.value_counts()

fig = plt.figure(figsize=(8,6))
ax = sns.boxplot(data=df,x="vehicleType",y="price")
ax.set(xlabel="Tipo de Veículo",ylabel="Range de Preço",title="Variação de preço por tipo de veículo")
plt.show()
```

## Contagem total de veículos à venda conforme o tipo de veículo

```python
fig = plt.figure(figsize=[10,8])

#df.vehicleType.hist()
df_vt = df.vehicleType.value_counts(sort=False)
#ax = sns.barplot(x=df_vt.index,y=df_vt,order=df.vehicleType.unique(),
 #           palette="BuPu")
ax = sns.countplot(df.vehicleType,palette="BuPu")
ax.set(xlabel="Tipo de Veículo",
      ylabel="Total de veículos para Venda",
      title="Contagem total de veículos à venda por tipo")
for p in ax.patches:
    ax.text(p.get_x()+p.get_width()/2.,p.get_height()+100,
            int(p.get_height()),ha="center")
plt.show()
```
